```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ["""Houve o aumento significativo dos
sintomas.""", """Não houve aumento ou diminuição de
tais sintomas.""", """Houve alguma diminuição dos sintomas"""]
sizes = [65, 25, 18]

fig, ax = plt.subplots()
ax.pie(sizes, autopct='%1.1f%%', colors = ["#808080", "#DCDCDC", "#F8F8FF"])


ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.legend(labels, 
          title ="Adaptação", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 

plt.show()
```
