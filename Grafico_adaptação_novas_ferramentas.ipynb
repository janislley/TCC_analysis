```python
import os
import numpy as np
from matplotlib import pyplot as plt


adaptacao = ["Fácil, não tive dificuldades no manuseio", 
             "Difícil, ainda estou me adaptando",
             "Moderado, com algumas dificuldades de manuseio"]
valores = [22, 26, 59]

fig, ax = plt.subplots()    
width = 0.5 # the width of the bars 
ind = np.arange(len(valores))  # the x locations for the groups
ax.barh(ind, valores, width, color="black")
ax.set_yticks(ind)
ax.set_yticklabels(adaptacao, minor=False)

plt.xlabel('Quantidade de pessoas.')

for i, v in enumerate(valores):
    ax.text(v, i, " "+str(v), color='black', va='center')

plt.show()

plt.savefig('books_read.png')
```
