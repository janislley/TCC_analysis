```python
import os
import numpy as np
from matplotlib import pyplot as plt


Pessoas = ["""Pessoa 5 (Sex M e 30 anos)""", 
          """Pessoa 8 (Sex F e 49 anos)""", 
          """Pessoa 15 (Sex F e 37 anos)""",
          """Pessoa 6 (Sex F e 42 anos)""", 
          """Pessoa 4 (Sex F e 35 anos)""", 
          """Pessoa 7 (Sex F e 32 anos)""",
          """Pessoa 16 (Sex F e 33 anos)""", 
          """Pessoa 3 (Sex F e 47 anos)""",
          """Pessoa 10 (Sex F e 56 anos)""",
          """Pessoa 12 (Sex F e 37 anos)""",
          """Pessoa 11 (Sex F e 39 anos)""",
          """Pessoa 14 (Sex F e 47 anos)""",
          """Pessoa 1 (Sex F e 27 anos)""",
          """Pessoa 2 (Sex F e 45 anos)""",
          """Pessoa 9 (Sex F e 49 anos)""",
          """Pessoa 13 (Sex F e 39 anos)"""]
valores = [3, 4, 5, 8, 8, 9, 10, 10, 13, 15, 15, 15, 18, 20, 20, 22]

fig, ax = plt.subplots()    
width = 0.7 # the width of the bars 
ind = np.arange(len(valores))  # the x locations for the groups
ax.barh(ind, valores, width, color="black")
ax.set_yticks(ind)
ax.set_yticklabels(Pessoas, minor=False)

plt.xlabel('Quantidade de sintomas de estresse.')

for i, v in enumerate(valores):
    ax.text(v, i, " "+str(v), color='black', va='center')

plt.show()

plt.savefig('books_read.png')
```
